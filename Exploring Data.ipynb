Listing Titles

```python
year = 2018
df = pd.read_csv(f"data/MHMS-{year}-{year+1}.csv", header=[0, 1], low_memory=False)
```

```python
df.columns.names = ["header_1", "header_2"]
```

```python
new_column_names = df.columns.to_frame(index=False, name=["header_1", "header_2"]).copy()
new_column_names.loc[new_column_names["header_1"].str.startswith("Unnamed"), "header_1"] = np.nan
new_column_names.at[0, "header_1"] = "No Group"
new_column_names["header_1"] = new_column_names["header_1"].ffill()

mask = new_column_names["header_2"].str.contains(r'\?', na=False)
new_column_names.loc[mask, "header_1"] = new_column_names.loc[mask, "header_2"]

new_column_names.head(n=100)
df.columns = pd.MultiIndex.from_frame(new_column_names)
```

```python
header_1 = df.columns.get_level_values("header_1")
q1 = list(header_1[~header_1.str.startswith("Unnamed")])
q1 = [item.strip() for item in q1]
```

```python
header_2 = df.columns.get_level_values("header_2")
q2 = list(header_2[~header_2.str.startswith("Unnamed")])    
q2 = [item.strip() for item in q2]
```

```python
with open("filtered_header_1.txt", "w") as f:
    for item in set(q1):
        f.write(f"{item.replace("\n", "")}\n")
```

```python
headerdict = {key: [] for key in q1}

for i in range(len(q1)):
    headerdict[q1[i]].append(q2[i])

with open("headers.txt", "w") as f:
    for key, value in headerdict.items():
        f.write(f"{key}\n\n")

        for q in value:
            if q != key:
                f.write(f"- {q}\n")

        f.write("\n\n============================\n\n")

headerdict
```

```python
themekeywords = {"physical": ["sports", "PE lessons", "physical"], 
          "mental": ["worried","enjoy"],
          "drugs": ["cigarette","drug","substance"], 
          "gamble": ["gamble"],
          "abuse": ["violence","bullying","hurtful"], 
          "sleep": ["sleep"],
          "family": ["family","death"],
          "food": [],
          "relationships": ["partner", "sexual"],
          "school": ["PSHE","lesson","college","school","learnt anything"],
          "other": []}


with open("themes.txt", "w") as f:
    for key, value in themekeywords.items():
        f.write(f"{key}\n\n")

        for keyword in value:
            for q in set(q1):
                if keyword in q:
                    f.write(f"- {q}\n")

        f.write("\n\n============================\n\n")

```
